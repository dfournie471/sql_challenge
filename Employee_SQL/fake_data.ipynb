```python
# import dependencies and create an engine for the Employee_SQL_DB database.
from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:postgres@localhost:5432/Employee_SQL_DB')
connection = engine.connect()
```

```python
#import pandas, matplotlib and other dependencies.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func
from sqlalchemy.ext.automap import automap_base
```

```python
# use inspect(engine) to list all the tables in the database.
inspector = inspect(engine)
inspector.get_table_names()
```

```python
# use engine.execute` to verify engine connection and display data from salaries.
engine.execute('SELECT * FROM salaries').fetchall()
```

```python
# Reflect Database into ORM class
Base = automap_base()
Base.prepare(engine, reflect=True)
Salaries = Base.classes.salaries
Employees = Base.classes.employees
Titles=Base.classes.titles
```

```python
 # Start a session to query Employee_SQL_DB
session = Session(bind=engine)
```

```python
# save queries from database into variables
results_1 = session.query(Salaries.salary,Salaries.employee_number).all()


results_2 = session.query(Employees.employee_title_id).all()
```

```python
#convert the tuples that the query variables output to lists
employee_id = [result[1] for result in results_1[:-1]]
salary = [result[0] for result in results_1[:-1]]
employee_title_id = [result[0] for result in results_2[:-1]]
```

```python
# created salary_df in order to plot with pandas.
salary_df=pd.DataFrame({'Salary':salary})
```

```python
# plot histogram
salary_hist=salary_df.hist(figsize=(15,8))
plt.xlabel("Salary Range",fontsize=16)
plt.ylabel("Number of Employees",fontsize=16)
plt.title("Employee Salary Histogram",fontsize=20)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlim((min(salary)-3000),(max(salary)))
plt.tight_layout()
plt.savefig("../output_data/Employee_salary_histogram.png")
plt.show()
```

```python
#create new df that includes salary and employee title ids
title_salary_df=pd.DataFrame({'Salary':salary,'Employee_title':employee_title_id})
```

```python
#group title_salary_df by Employee_title
grouped_by_title=title_salary_df.groupby('Employee_title')
```

```python
#find the mean of the groups
salary_average= grouped_by_title.mean()
```

```python
#plot the bar graph
job_list=['Assistant Engineer','Engineer','Senior Engineer','Technique Leader','Manager','Staff', 'Senior Staff']
ax=salary_average.plot(kind='bar',figsize=(15,8),facecolor='green',alpha=.67,rot=0, legend=False)
ax.set_xlabel('Employee Title', fontsize=16)
ax.set_ylabel('Average Salary', fontsize=16)
ax.set_title("Average Salary for each Job Title", fontsize=20)
ax.set_xticklabels(job_list)
ax.tick_params(axis="x", labelsize=13)
ax.tick_params(axis="y", labelsize=14)

for i in ax.patches:
    ax.text(i.get_x()+.04, i.get_height()+1000, \
            str(round((i.get_height()), 2)), fontsize=11.5, color='black',
                rotation=0)
plt.savefig("../output_data/Average_Salary.png")
plt.show()
```
